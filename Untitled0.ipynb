```
#방법2
import requests
from bs4 import BeautifulSoup

urls = []
url = 'https://ridibooks.com/category/bestsellers/2200?page='
for i in range(1, 6):
    urls.append(url + str(i))

    data = []
    ranking= []
for url in urls:
    response = requests.get(url)
    response.encoding = 'utf-8'
    html = response.text

    soup = BeautifulSoup(html, 'html.parser')

    bookservices = soup.select('.title_text')     # col-lg-6 클래스 안의 h2 태그 탐색
    bookranking= soup.select('.book_ranking')
    for book in bookservices:
        data.append(book.text.strip())
    for rank in bookranking:
        ranking.append(rank.text.strip())

for no, book in enumerate(data, 1):
    print(no, book)
```
